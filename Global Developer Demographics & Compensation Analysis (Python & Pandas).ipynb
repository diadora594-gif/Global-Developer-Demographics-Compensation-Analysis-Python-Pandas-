## 🔧 Imports and Data Loading

```python
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

path = '/content/drive/MyDrive/Colab Notebooks/survey_results_public.csv'

pathSchema = '/content/drive/MyDrive/Colab Notebooks/survey_results_schema.csv'

df = pd.read_csv(path, encoding = 'utf-8')

schema = pd.read_csv(pathSchema, encoding = 'utf-8')


df.head()
```

## ✅ Task 1 — Total Respondents

```python
n = df['ResponseId'].nunique()
print(f"Total respondents: {n:,}")
```

## ✅ Task 2 — Respondents with Full Answers

```python
questions = set(schema['qname'])
columns = set(df.columns)

common_questions = list(questions.intersection(columns))
df_filtered = df[common_questions]

df_clean = df_filtered[df_filtered.notna().all(axis=1)]

print(f"Questions used: {len(common_questions)}")
print(f"Respondents with full answers: {len(df_clean):,}")
```

## ✅ Task 3 — Work Experience Statistics

```python
stats = pd.Series({
    'Mean': round(df['WorkExp'].mean(),2),
    'Median': df['WorkExp'].median(),
    'Mode': df['WorkExp'].mode()[0]
})

print(stats)
```

## ✅ Task 4 — Remote Workers

```python
remote_count = df[df['RemoteWork'] == 'Remote'].shape[0]
print(f"Remote workers: {remote_count:,}")
```

## ✅ Task 5 — Python Popularity

```python
python_users = df['LanguageHaveWorkedWith'].str.contains('Python', case = False, na=False)
percentage = python_users.mean() * 100

print(f"Python users: {percentage:.2f}%")


```

## ✅ Task 6 — Online Learning

```python
online_learners = df['LearnCode'].str.contains('Online', case=False, na=False)
count_online = online_learners.sum()

print(f"Online learners: {count_online:,}")
```

## ✅ Task 7 — Python Developers Compensation by Country

```python
df['is_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', case = False, na=False)

python_df = df[df['is_python']]

comp = python_df.groupby('Country')['ConvertedCompYearly'].agg(['mean', 'median']).dropna()
comp = comp.sort_values(by='mean', ascending=False)

comp.head()
```

## ✅ Task 8 — Top 5 Highest Paid Education Levels

```python
top5 = df.dropna(subset=['ConvertedCompYearly']) \
    .sort_values(by='ConvertedCompYearly', ascending=False) \
    .head(5).reset_index()

top5[['EdLevel', 'ConvertedCompYearly']]


```

## ⭐ Task 9 — Python by Age

```python
df['is_python'] = df['LanguageHaveWorkedWith'].str.contains('Python', case = False, na=False)

age_python = df.groupby('Age')['is_python'].mean() * 100
age_python.sort_values(ascending=False)
```

## ⭐ Task 10 — Top Industries (High-paid Remote)

```python
threshold = df['ConvertedCompYearly'].quantile(0.75)

filtered = df[
    (df['ConvertedCompYearly'] > threshold) &
    (df['RemoteWork'] == 'Remote')
]

industries = filtered['Industry'].value_counts()
industries.head()

```
